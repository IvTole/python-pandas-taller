## Filtrados con más de una condición

```python
employees_df = pd.read_csv("../Datasets/employees.csv", parse_dates=["Start Date"])
employees_df["Last Login Time"] = pd.to_datetime(employees_df["Last Login Time"], format="%H:%M %p").dt.time
employees_df["Senior Management"] = employees_df["Senior Management"].astype(bool)
employees_df["Gender"] = employees_df["Gender"].astype("category")
employees_df.head()
```

Con otros operadores 

* `&` (AND)
* `|` (OR)
* `~` (NOT)


```python
is_senior_management = employees_df["Senior Management"] # esta variable ya es un boolean
started_in_80s = employees_df["Start Date"] < "1990-01-01"

# empleados en senior management O que hayan empezado antes de los 90`s

employees_df[is_senior_management | started_in_80s]
```

```python
is_robert = employees_df["First Name"] == "Robert"
is_in_client_services = employees_df["Team"] == "Client Services"
start_date_after_june_2016 = employees_df["Start Date"] > "2016-06-01"

# empleado cuyo nombre sea "Robert" Y que se encuentre en el departamenteo de servicio al cliente
# O que haya empezado después de junio de 2016

employees_df[(is_robert & is_in_client_services) | start_date_after_june_2016]
```

## Método isin

* Regresa True o False para una fila si algún valor se encuentra en una lista/tupla/serie que uno le dé de referencia.

```python
employees_df = pd.read_csv("../Datasets/employees.csv", parse_dates=["Start Date"])
employees_df["Last Login Time"] = pd.to_datetime(employees_df["Last Login Time"], format="%H:%M %p").dt.time
employees_df["Senior Management"] = employees_df["Senior Management"].astype(bool)
employees_df["Gender"] = employees_df["Gender"].astype("category")
employees_df.head()
```

```python
# Departamento Legal, de Ventas, O de Productos

legal_team = employees_df["Team"] == "Legal"
sales_team = employees_df["Team"] == "Sales"
product_team = employees_df["Team"] == "Product"

employees_df[legal_team | sales_team | product_team]
```

```python
# Este mismo dataframe también puede crearse a través del metodo is in
target_teams = ["Legal", "Sales", "Product"]
series_teams = employees_df["Team"].isin(target_teams)

employees_df[series_teams]
```

## Método duplicated y drop_duplicates

duplicated
* Regresa un True si el valor en la serie es un duplicado
* La primera ocurrencia se marca como False, pero si lo vuelve a encontrar sería True.
* Se puede utilizar el parámetro **keep** para designar si quedarse con la primera o la última ocurrencia.
* **keep = False** sería para marcar todos los valores repetidos como True.
* Recordar que se puede utilizar el símbolo **~** para invertir True-False.


drop_duplicates
* elimina las filas con valores repetidos.
* por default elimina solo si TODOS sus valores se repite.
* El parámetro **subset** nos permite modificar esto último.

```python
employees_df = pd.read_csv("../Datasets/employees.csv", parse_dates=["Start Date"])
employees_df["Last Login Time"] = pd.to_datetime(employees_df["Last Login Time"], format="%H:%M %p").dt.time
employees_df["Senior Management"] = employees_df["Senior Management"].astype(bool)
employees_df["Gender"] = employees_df["Gender"].astype("category")
employees_df.head()
```

```python
employees_df[employees_df["First Name"].duplicated()]
```

```python
employees_df[employees_df["First Name"].duplicated(keep="first")]
```

```python
employees_df[employees_df["First Name"].duplicated(keep="last")]
```

```python
employees_df[employees_df["First Name"].duplicated(keep=False)]
```

```python
employees_df[~employees_df["First Name"].duplicated(keep=False)]
```

```python
employees_df.drop_duplicates()

```

```python
employees_df.drop_duplicates("Team")

```

```python
employees_df.drop_duplicates("Team", keep="first")

```

```python
employees_df.drop_duplicates("Team", keep="last")

```

```python
employees_df.drop_duplicates("Team", keep=False)

```

```python
employees_df.drop_duplicates("First Name", keep=False)

```

```python
employees_df.drop_duplicates(["Senior Management", "Team"]).sort_values("Team")

```

```python
employees_df.drop_duplicates(["Senior Management", "Team"], keep="last").sort_values("Team")
```
