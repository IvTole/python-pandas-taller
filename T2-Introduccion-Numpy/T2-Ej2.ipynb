```python
# Librerías utlizadas

import numpy as np # operaciones de arrays
import matplotlib.pyplot as plt # gráficos
import csv # manejo de archivos csv

%matplotlib inline
```

```python
# 1. Leer las primeras 5 líneas y guardarlas en una lista

csvfile= open("../Datasets/31001_test.csv","r") # Se abre el archivo 31001.csv con permisos de lectura
lector  = csv.reader(csvfile, delimiter=',') # Se asigna a una variable
print ("Lectura del  fichero  con modulo csv\n")
```

```python
# pero solo quiero las primeras 5 líneas

csvfile= open("../Datasets/31001_test.csv","r")
lector  = csv.reader(csvfile, delimiter=',')
print ("Lectura del  fichero  con modulo csv\n")

lineas_n = 5 # numero de lineas solicitadas para leer
lineas_iteracion = 0 # se inicializa indice de la iteracion sobre las lineas
lineas_leidas = [] # se inicializa la lista donde se le agrega (append) cada lista de observaciones de temperatura

for linea in lector:
    lineas_iteracion += 1 # se agrega 1 al contador
    lineas_leidas.append(linea) # se agrega el elemento leido a a la lista
    if lineas_iteracion == lineas_n: # condicion de termino del loop, solo queremos 5 lineas, no todas
        break # el comando break saca del loop, el comando continue continua con la siguiente iteracion

csvfile.close() # se cierra el archivo para que no gaste recursos

# la lista "lineas_leidas" incluye los 5 primeros renglones (que son listas también). Es una lista de listas.
print('Lista con las primeras ',lineas_n,' lineas = \n', lineas_leidas) # 
```

```python
print('longitud de la primera linea = ', len(lineas_leidas[0]))
print('longitud de la segunda linea = ', len(lineas_leidas[1]))
print('longitud de la tercera linea = ', len(lineas_leidas[2]))
print('longitud de la cuarta linea = ', len(lineas_leidas[3]))
print('longitud de la quinta linea = ', len(lineas_leidas[4]))
```

La longitud de los elementos es la misma (34), sin embargo, observamos que a veces hay elementos vacíos. Se observa que las observaciones no tienen el mismo número de elementos. Supongo que son medidas de temperatura por cada mes, y que cada mes tiene diferente número de días. Además, cada elemento se guardó como un string.

```python
# creamos una copia de las lista original para modificarla
lineas_leidas_t = lineas_leidas.copy() 

for i in range(0,len(lineas_leidas_t)):
    #observation = observation.replace('','nan') # es una lista, aplicamos replace
    lineas_leidas_t[i] = np.array(lineas_leidas_t[i]) # se hace un cambio de list a array en cada elemento
    #lineas_leidas_t[i][lineas_leidas_t[i]==''] = 'nan'
    lineas_leidas_t[i] = np.where(lineas_leidas_t[i]=='','nan',lineas_leidas_t[i])
    # el método where de numpy permite hacer una sustitución basado en una condición np.where(condition, replace if true, replace if false)
    lineas_leidas_t[i] = lineas_leidas_t[i][3:].astype('float') # sólo los elementos después del índice 3 pueden cambiar a valor float
lineas_leidas_t = np.array(lineas_leidas_t)  # se hace un array de toda la lista resultante

# imprimimos un ejemplo
lineas_leidas_t[3]
```

```python
# a los objetos numpy array podemos extraer directamente la media

lineas_leidas_t[2].mean()
```

```python
lineas_leidas_t[3].mean()
# vemos que hay problemas cuando hay valores nulos
```

```python
# 3. Obtener dato minimo, maximo y promedio de cada observacion.

# Usamos un loop para recorrer cada observacion/array de temperaturas de cada mes

observacion = 0 # se inicializa indice de la observacion de temperaturas

# Se inicializan las listas donde se guarda la informacion para los graficos
lista_obs = []
lista_minimo = []
lista_maximo = []
lista_promedio = []
lista_std = []

# listas para las barras de error
lista_errmax = []
lista_errmmin = []


for obs in lineas_leidas_t:
    line = obs[~np.isnan(obs)] ## Se crea una linea donde se remuven los valores NaN. Aqui es donde se hace la estadistica
    observacion += 1 # se aumenta 1 el numero de observacion
    lista_obs.append(observacion)
    minimo = np.min(line)
    lista_minimo.append(minimo)
    maximo = np.max(line)
    lista_maximo.append(maximo)
    promedio = np.mean(line)
    lista_promedio.append(promedio)
    std = np.std(line)
    lista_std.append(std)
    
    lista_errmax.append(maximo - promedio) # listas para las barras de error 
    lista_errmmin.append(promedio - minimo)
    print('La observacion #', observacion,' tiene un minimo de ', minimo, ', un maximo de ', maximo, ', y un promedio de ', promedio, ' grados Celsius')

```

```python
# Graficamos los promedios de cada día con su barra de error

import matplotlib.pyplot as plt
import numpy as np

plt.errorbar(lista_obs, lista_promedio, std, fmt='ok', lw=3)
plt.errorbar(lista_obs, lista_promedio, [lista_errmmin, lista_errmax], fmt = '.k', ecolor = 'gray', lw = 1)
plt.xlabel('# de observacion')
plt.ylabel('Temperatura promedio [Cº]')

plt.show()
```

## Alternativa, con un objetos DataFrame de Pandas

```python
import pandas as pd
```

```python
df = pd.read_csv("../Datasets/31001_test.csv", header=None)
df.head(5)
# El dataset no contiene un header para las columnas
```

Observamos que automáticamente los valores vacíos los pone como ``Nan``. Ya no es necesarios hacer los primero pasos.

```python
# Podemos extraer directamente las primeras 5 filas
# Además, podemos reestringir también las columnas, quedándonos a partir de la columna 3
df_filtered = df.iloc[0:5,3:] # recordar que el índice empieza en 0
df_filtered
```

```python
# Ya tenemos nuestros cinco registros, solo tenemos que hacer la estadística por fila

for index in df_filtered.index:
    media = df_filtered.iloc[index].mean()
    print("Media de temperatura en observación número ", index, ' es ', media)
```
