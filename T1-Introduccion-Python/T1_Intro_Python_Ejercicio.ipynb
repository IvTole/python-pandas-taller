```python
df = pd.read_csv("../Datasets/netflix_titles.csv")
df.head(5)  # Mostrar las primeras 5 líneas
```

La pregunta a resolver es la siguiente:

### Miembros del casting con mayor número de contenido.

```python
# Podemos observar que tenemos ciertos valores nulos que nos podría causar problemas
df_filtered_cast = df[df.cast.notnull()]
df_filtered_cast.head(20)
```

```python
# Extraemos una lista de toda esta columna "Cast"
list_cast = df_filtered_cast['cast'].values.tolist()
list_cast
```

```python
# Vemos como están los elementos de esta lista
list_cast[0]
```

```python
# Vemos que el formato permite realizar una separación utilizando ", " como separados (coma y espacio)
# Por ejemplo, al primer elemento de la lista cast

list_cast[0].split(sep=', ')

# y de esta forma tenemos separados a los elementos de este índice 0
```

```python
# realizamos esta operacion para cada uno de los elemento de la lista de cast
# lo añadimos (append) sobre una nueva variable list_cast_split, que comenzamos vacía

list_cast_split = [] # comenzamos una nueva lista sin elementos

for i in range(0,len(list_cast)):
    
    element_split = list_cast[i].split(", ") # hacemos la separacion por cada elemento
    list_cast_split.append(element_split) # lo agregamos a una nueva lista 
    
list_cast_split
```

```python
# Tenemos un formato más adecuado, pero todavía tenemos lista de listas
# Realizamos lo que se llama list flatten

list_cast_split_flatten = [element for sublist in list_cast_split for element in sublist]
list_cast_split_flatten
```
